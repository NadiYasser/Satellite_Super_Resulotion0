```python
%load_ext autoreload
%autoreload 2
import os 
os.chdir("/home/yasser/classes/DL26/final_Project")

import torch 
import matplotlib.pyplot as plt
from PIL import Image
```

```python

# Manually chosen image names
image_names = [
    "austin16_r00_c04.png",
    "vienna25_r10_c15.png",
    "img_12.png",
    "img_23.png",
    "img_31.png"
]

lr_dir = "data/preprocessed/test/LR"
hr_dir = "data/preprocessed/test/HR"

# Create figure
fig, axes = plt.subplots(len(image_names), 2, figsize=(8, 15))

for i, img_name in enumerate(image_names):
    lr_img = Image.open(f"{lr_dir}/{img_name}")
    hr_img = Image.open(f"{hr_dir}/{img_name}")

    axes[i, 0].imshow(lr_img)
    axes[i, 0].set_title("LR")
    axes[i, 0].axis("off")

    axes[i, 1].imshow(hr_img)
    axes[i, 1].set_title("HR (Ground Truth)")
    axes[i, 1].axis("off")

plt.tight_layout()
plt.show()
```
