```python
# ============================================================
# Unified Training & Validation Curves (Paper Style)
# SRCNN vs SRResNet vs EDSR
# ============================================================

import json
from pathlib import Path
import matplotlib.pyplot as plt

# ------------------------------------------------------------
# Base directory (notebooks/said/)
# ------------------------------------------------------------
BASE_DIR = Path.cwd()

# ------------------------------------------------------------
# Paths to JSON training histories
# ------------------------------------------------------------
PATHS = {
    "SRCNN": BASE_DIR / "../../src/models/checkpoints/SRCNN/srcnn_training_history.json",
    "SRResNet": BASE_DIR / "../../src/models/checkpoints/SRRESNET/srresnet.json",
    "EDSR": BASE_DIR / "../../src/models/checkpoints/EDSR/edsr_training_history0.json",
}

# ------------------------------------------------------------
# Load history
# ------------------------------------------------------------
def load_history(path):
    with open(path, "r") as f:
        return json.load(f)

srcnn = load_history(PATHS["SRCNN"])
srresnet = load_history(PATHS["SRResNet"])
edsr = load_history(PATHS["EDSR"])

# ------------------------------------------------------------
# Extract metrics (your exact JSON structure)
# ------------------------------------------------------------
data = {
    "SRCNN": {
        "train_loss": srcnn["train_losses"],
        "val_loss": srcnn["val_losses"],
        "train_psnr": srcnn["train_psnrs"],
        "val_psnr": srcnn["val_psnrs"],
    },
    "SRResNet": {
        "train_loss": srresnet["train_losses"],
        "val_loss": srresnet["val_losses"],
        "train_psnr": srresnet["train_psnrs"],
        "val_psnr": srresnet["val_psnrs"],
    },
    "EDSR": {
        "train_loss": edsr["train_losses"],
        "val_loss": edsr["val_losses"],
        "train_psnr": edsr["train_psnrs"],
        "val_psnr": edsr["val_psnrs"],
    },
}

# ------------------------------------------------------------
# Colors (fixed per model)
# ------------------------------------------------------------
COLORS = {
    "SRCNN": "#1f77b4",     # blue
    "SRResNet": "#ff7f0e",  # orange
    "EDSR": "#2ca02c",      # green
}

# ------------------------------------------------------------
# Create figure
# ------------------------------------------------------------
plt.figure(figsize=(14, 6))

# ============================================================
# Subplot 1: Loss (Train + Validation)
# ============================================================
plt.subplot(1, 2, 1)

for model, values in data.items():
    epochs_train = range(1, len(values["train_loss"]) + 1)
    epochs_val = range(1, len(values["val_loss"]) + 1)

    # Validation (continuous line)
    plt.plot(
        epochs_val,
        values["val_loss"],
        color=COLORS[model],
        linewidth=2,
        label=f"{model} (Val)"
    )

    # Training (points)
    plt.plot(
        epochs_train,
        values["train_loss"],
        linestyle="None",
        marker="o",
        markersize=3,
        alpha=0.6,
        color=COLORS[model],
        label=f"{model} (Train)"
    )

plt.xlabel("Epoch")
plt.ylabel("Loss (L1)")
plt.title("Training and Validation Loss Comparison")
plt.grid(True, linestyle="--", alpha=0.5)
plt.legend()

# ============================================================
# Subplot 2: PSNR (Train + Validation)
# ============================================================
plt.subplot(1, 2, 2)

for model, values in data.items():
    epochs_train = range(1, len(values["train_psnr"]) + 1)
    epochs_val = range(1, len(values["val_psnr"]) + 1)

    # Validation (continuous line)
    plt.plot(
        epochs_val,
        values["val_psnr"],
        color=COLORS[model],
        linewidth=2,
        label=f"{model} (Val)"
    )

    # Training (points)
    plt.plot(
        epochs_train,
        values["train_psnr"],
        linestyle="None",
        marker="o",
        markersize=3,
        alpha=0.6,
        color=COLORS[model],
        label=f"{model} (Train)"
    )

plt.xlabel("Epoch")
plt.ylabel("PSNR (dB)")
plt.title("Training and Validation PSNR Comparison")
plt.grid(True, linestyle="--", alpha=0.5)
plt.legend()

# ------------------------------------------------------------
# Save & show
# ------------------------------------------------------------
plt.tight_layout()
plt.savefig("unified_train_val_comparison_SRCNN_SRResNet_EDSR.png", dpi=300)
plt.show()

print(" Unified train/validation comparison figure generated successfully.")
```
